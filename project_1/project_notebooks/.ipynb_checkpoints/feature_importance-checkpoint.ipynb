```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import xgboost as xgb

import joblib

plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['figure.figsize'] = 10, 6
```

```python
data_path = "/Users/oliverpaul/Data_Science/idiap/lucideles/data"
```

```python
def mae(y, yhat):
    """
    Computes Mean Absolute Error
    """
    return np.sum(np.abs(np.array(y) - np.array(yhat))) / len(np.array(y))
```

```python
# loading data
train = np.load(data_path + '/train.npy') 
valid = np.load(data_path + '/valid.npy') 

train_x, train_y = train[:,:-2], train[:,-2:]
valid_x, valid_y = valid[:,:-2], valid[:,-2:]
```

```python
#loading xgboost models
xg_model_dgp = joblib.load(data_path + '/xg_model_dgp')
xg_model_ill = joblib.load(data_path + '/xg_model_ill')
```

```python
def permutation_importance(model, x, y, target):
    """
    Takes features and target and calculates permutation importance
    by predicting y for each shuffled feature in turn. 'azimuth_sin'
    and 'azimuth_cos' are highly correlated so we drop both these columns 
    together.
    Args:
        x: Validation set
        y: True target 
        target: name of target. either 'dgp' or 'ill'
    """
    
    if target == 'dgp':
        target_index = 0
    else:
        target_index = 1
        
    x = x.copy()
    y = y.copy()
        
    X_XGB = xgb.DMatrix(x, label=y[:,target_index].copy())
    
    model = model
    #getting baseline mae score for fully trained XGB model
    baseline = mae(model.predict(X_XGB), y[:,target_index])
    imp = []
    for i in range(x.shape[1]):
        #indexes of azimuth_sin and cos are 5 and 6 respectively
        if i == 5:
            save = x[:,[5, 6]].copy()
            x[:,[5, 6]] = np.random.permutation(x[:,[5, 6]])
            X_XGB = xgb.DMatrix(x, label=y[:,target_index].copy())
            loss = mae(model.predict(X_XGB), y[:,target_index])
            x[:,[5, 6]] = save
            imp.append(abs(baseline - loss))
            break
        save = x[:,i].copy()
        x[:,i] = np.random.permutation(x[:,i])
        X_XGB = xgb.DMatrix(x, label=y[:,target_index].copy())
        loss = mae(model.predict(X_XGB), y[:,target_index])
        x[:,i] = save
        imp.append(abs(baseline - loss))
    return np.array(imp)
```

```python
per_imp_dgp = permutation_importance(xg_model_dgp, valid_x, valid_y, 'dgp')
per_imp_ill = permutation_importance(xg_model_ill, valid_x, valid_y, 'ill')
```

```python
def plot_permutation_importance(imps, features, target):
    
    features = features
    imp = pd.Series(imps, index=features)
    sorted_ix = imp.sort_values().index
    ax = imp[sorted_ix].plot(kind='barh', width = .7, color='royalblue', alpha=0.9)
    ax.set_title('Permutation importance ' + target)
    ax.set_xlabel('Change in MAE')
    ax.grid(color='k', ls = '-.', lw = 0.2, axis='x')
    plt.show()
```

```python
features = ['blind_angle', 'fully_open', 'ibn', 'idh', 'altitude', 'azimuth']

plot_permutation_importance(per_imp_dgp, features, 'Glare Probability')
```

```python
plot_permutation_importance(per_imp_ill, features, 'Illuminance')
```
